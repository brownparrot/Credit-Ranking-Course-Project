## helper Functions 

```python
def count_strings(val):
    return 0 if pd.isna(val) else len(val.split(','))
```

```python
def convert_to_months(val):
    if pd.isna(val):
        return 0
    else:
        years, months = int(val.split(" ")[0]), int (val.split(" ")[3])
    return years * 12 + months
```

```python
def clean_and_convert_to_numeric(df, columns):
    """
    Cleans and converts specified columns to numeric.
    Non-numeric characters are removed, and the values are converted to numeric.
    """
    for col in columns:
        if df[col].dtype == 'object':
            df[col] = df[col].str.replace(r'[^-.0-9]', '', regex=True)
        df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
def fill_missing_values(df, columns, group_by_col=None):
    """
    Fills missing values in specified columns. 
    Optionally, fills based on the median value grouped by a column.
    """
    for col in columns:
        if group_by_col:
            df[col] = df.groupby(group_by_col)[col].transform(lambda x: x.fillna(x.median()))
        df[col].fillna(df[col].median(), inplace=True)
```

```python
def impute_with_group_mode(df, column, group_by_col, test_df=None):
    """
    Imputes missing values in a column based on the mode within a specified group.
    If no group mode exists, it uses the overall mode.
    Optionally applies the same logic to a test DataFrame.
    """
    # Calculate the overall mode
    overall_mode = df[column].mode()[0]

    # Impute missing values with group mode or overall mode in the primary DataFrame
    df[column] = df.groupby(group_by_col)[column].transform(
        lambda x: x.fillna(x.mode()[0] if not x.mode().empty else overall_mode)
    )
    df[column].fillna(overall_mode, inplace=True)

    # Optionally handle the test DataFrame
    if test_df is not None:
        test_df[column] = test_df.groupby(group_by_col)[column].transform(
            lambda x: x.fillna(x.mode()[0] if not x.mode().empty else overall_mode)
        )
        test_df[column].fillna(overall_mode, inplace=True)
        return df, test_df

    return df
```

```python
def impute_categorical_columns_with_issues(df, issue_dict, group_by_col):
    """
    Imputes problematic values in categorical columns based on the mode within groups.
    If no group mode exists, it falls back to the overall mode.
    """
    for col, problem_value in issue_dict.items():
        if col in df.columns:
            # Calculate the overall mode, excluding problematic values
            overall_mode = df.loc[df[col] != problem_value, col].mode().get(0, 'Unknown')

            # Replace problematic values with group mode or overall mode
            df[col] = df.groupby(group_by_col)[col].transform(
                lambda x: x.replace(problem_value, x[x != problem_value].mode().get(0, overall_mode))
            )

            # Log results for verification (optional)
            print(f"Value counts for {col}:\n", df[col].value_counts())

    return df
```

## Remove Outliers

```python
def remove_outliers(df, columns, iqr_multiplier=1.5):
    """
    Removes outliers from the specified columns of a DataFrame based on the IQR method.
    
    """
    for col in columns:
        if df[col].dtype in ['int64', 'float64']:
            # Calculate Q1 (25th percentile) and Q3 (75th percentile)
            Q1 = df[col].quantile(0.01)
            Q3 = df[col].quantile(0.99)
            IQR = Q3 - Q1

            # Define outlier bounds
            lower_bound = Q1 - iqr_multiplier * IQR
            upper_bound = Q3 + iqr_multiplier * IQR

            # Filter out outliers
            df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    
    return df

# List of numeric columns
numeric_cols = [
    'Age', 'Income_Annual', 'Base_Salary_PerMonth', 'Total_Bank_Accounts', 
    'Total_Credit_Cards', 'Rate_Of_Interest', 'Total_Current_Loans', 
    'Delay_from_due_date', 'Total_Delayed_Payments', 'Credit_Limit', 
    'Total_Credit_Enquiries', 'Current_Debt_Outstanding', 
    'Ratio_Credit_Utilization', 'Per_Month_EMI', 'Monthly_Investment', 
    'Monthly_Balance', 'Loan_Count'
]

# Remove outliers from df_train
df_train = remove_outliers(df_train, numeric_cols)
```
